```python
from unidecode import unidecode

# Function to convert player names to proper English characters
def clean_player_names(name):
    return unidecode(name)

# List of DataFrames with player names that need cleaning
player_dataframes = [
    ('Transformed Datasets/player_appear_totals.csv', 'player_appear_totals'),
    ('Transformed Datasets/player_goal_totals.csv', 'player_goal_totals'),
    ('Transformed Datasets/player_appear_details.csv', 'player_appear_details'),
    ('Transformed Datasets/player_goal_details.csv', 'player_goal_details'),
    ('Transformed Datasets/top_goal_scorer.csv', 'top_goal_scorer'),
    ('Transformed Datasets/player_total_appearances.csv', 'player_total_appearances')
]

# Apply cleaning to all player dataframes
for file_path, df_name in player_dataframes:
    try:
        df = pd.read_csv(file_path)
        
        # Clean player names (usually in 'Player' column)
        if 'Player' in df.columns:
            df['Player'] = df['Player'].apply(clean_player_names)
            df.to_csv(file_path, index=False)
            print(f"✓ Cleaned {df_name}: Updated player names")
        else:
            print(f"✗ {df_name}: No 'Player' column found")
    except FileNotFoundError:
        print(f"✗ {df_name}: File not found")
    except Exception as e:
        print(f"✗ {df_name}: Error - {str(e)}")
```

```python
# Clean club and coach names across all datasets
from unidecode import unidecode

def clean_names(name):
    """Convert names with non-English characters to proper English"""
    return unidecode(str(name))

# Files with 'Club' column
club_files = [
    'Transformed Datasets/player_appear_details.csv',
    'Transformed Datasets/player_goal_details.csv',
    'Transformed Datasets/top_goal_scorer.csv',
    'Transformed Datasets/all_time_ranking_by_club.csv',
    'Transformed Datasets/coaches_appear_details.csv'
]

# Files with 'Coach' column
coach_files = [
    'Transformed Datasets/coaches_appear_details.csv',
    'Transformed Datasets/coaches_appear_totals.csv'
]

# Clean Club names
print("Cleaning Club Names:")
for file_path in club_files:
    try:
        df = pd.read_csv(file_path)
        if 'Club' in df.columns:
            df['Club'] = df['Club'].apply(clean_names)
            df.to_csv(file_path, index=False)
            print(f"  ✓ {file_path.split('/')[-1]}: Club names cleaned")
        else:
            print(f"  - {file_path.split('/')[-1]}: No 'Club' column")
    except FileNotFoundError:
        print(f"  ✗ {file_path}: File not found")
    except Exception as e:
        print(f"  ✗ {file_path}: Error - {str(e)}")

# Clean Coach names
print("\nCleaning Coach Names:")
for file_path in coach_files:
    try:
        df = pd.read_csv(file_path)
        if 'Coach' in df.columns:
            df['Coach'] = df['Coach'].apply(clean_names)
            df.to_csv(file_path, index=False)
            print(f"  ✓ {file_path.split('/')[-1]}: Coach names cleaned")
        else:
            print(f"  - {file_path.split('/')[-1]}: No 'Coach' column")
    except FileNotFoundError:
        print(f"  ✗ {file_path}: File not found")
    except Exception as e:
        print(f"  ✗ {file_path}: Error - {str(e)}")

print("\n✓ All names cleaned successfully!")
```

```python
# Standardize club names across all datasets
# Create a mapping from various club name variations to the standard name

club_name_mapping = {
    # Manchester United variations
    'Manchester United FC': 'Manchester United',
    
    # Other potential variations (add as needed)
    'Bayern Munich': 'FC Bayern Munchen',
    'Bayern Munchen': 'FC Bayern Munchen',
    'Barcelona': 'FC Barcelona',
    'Real Madrid': 'Real Madrid CF',
    'Inter Milan': 'FC Internazionale Milano',
    'Internazionale': 'FC Internazionale Milano',
    'Atletico Madrid': 'Club Atletico de Madrid',
    'PSG': 'Paris Saint-Germain',
    'Lyon': 'Olympique Lyonnais',
    'Porto': 'FC Porto',
    'Ajax': 'AFC Ajax',
    'Benfica': 'SL Benfica',
    'Monaco': 'AS Monaco FC',
    'Schalke 04': 'FC Schalke 04',
    'Schalke': 'FC Schalke 04',
}

def standardize_club_name(club_name):
    """Standardize club names using the mapping"""
    return club_name_mapping.get(club_name, club_name)

# Files that contain club names
files_with_clubs = [
    'Transformed Datasets/player_appear_details.csv',
    'Transformed Datasets/player_goal_details.csv',
    'Transformed Datasets/top_goal_scorer.csv',
    'Transformed Datasets/all_time_ranking_by_club.csv',
    'Transformed Datasets/coaches_appear_details.csv',
    'Transformed Datasets/coaches_appear_totals.csv'
]

print("Standardizing Club Names:")
for file_path in files_with_clubs:
    try:
        df = pd.read_csv(file_path)
        if 'Club' in df.columns:
            # Store original unique clubs for comparison
            original_clubs = df['Club'].nunique()
            
            # Apply standardization
            df['Club'] = df['Club'].apply(standardize_club_name)
            
            # Save back
            df.to_csv(file_path, index=False)
            
            new_clubs = df['Club'].nunique()
            print(f"  ✓ {file_path.split('/')[-1]}: {original_clubs} → {new_clubs} unique clubs")
        else:
            print(f"  - {file_path.split('/')[-1]}: No 'Club' column")
    except FileNotFoundError:
        print(f"  ✗ {file_path}: File not found")
    except Exception as e:
        print(f"  ✗ {file_path}: Error - {str(e)}")

print("\n✓ Club name standardization complete!")
```
